```
from bs4 import BeautifulSoup
```

```
import requests
```

```
doc=requests.get('https://medium.com/topics')
```

```
doc.content
```

```
bs=BeautifulSoup(doc.content)
```

```
bs.prettify()
```

```
link_dev=bs.find_all('div',class_="u-flex1 u-flexCenter u-paddingLeft20 u-paddingRight20")
```

```
topics=[]
for div_link in link_dev:
  print(div_link.find('a').text)
  topics.append([div_link.find('a').text,div_link.find('a')['href']])
```

```
topics
```

```
import pandas as pd
```

```
topics_data=pd.DataFrame(topics,columns=['Topic Name','Topic Link'])
```

```
topics_data.head()
```

```
# Save the dataframe to CSV File

topics_data.to_csv('topic_names.csv',index_label=False)
```
